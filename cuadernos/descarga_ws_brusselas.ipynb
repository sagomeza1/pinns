```python
# Standard library
import datetime as dt
from pathlib import Path

# Third-party
import numpy as np
import pandas as pd
import requests
import scipy.io as sio
```

```python
DATA_RAW_DIR = Path().cwd().parent / "data" / "raw"
```

```python
url = "https://zenodo.org/records/10723261/files/Weather_data.mat"
path_data = DATA_RAW_DIR / "weather_data.mat"
```

```python
response = requests.get(url)
# Verificar que la descarga fue exitosa
if response.status_code == 200:
    with open(path_data, "wb") as f:
        f.write(response.content)
    print(f"Archivo guardado como: {path_data}")
    WS_data = sio.loadmat(path_data)
    print("Se carga el archivo .mat")
else:
    print(f"Error al descargar: {response.status_code}")
```

```python
weather_data_dict = dict()
WS_data.keys()
```

```python
date = WS_data["Date"]
date = date[0]
date = [dt[0] for dt in date]
weather_data_dict["Date"] = pd.to_datetime(date)
```

```python
weather_data_dict["Lon"] = WS_data["Lon"][0]
weather_data_dict["Lat"] = WS_data["Lat"][0]
weather_data_dict["Alt"] = WS_data["Alt"][0]
weather_data_dict["Temperature"] = WS_data["Temperature"][0]
weather_data_dict["WindSpeed"] = WS_data["WindSpeed"][0]
weather_data_dict["WindDirectionX"] = WS_data["WindDirectionX"][0]
weather_data_dict["WindDirectionY"] = WS_data["WindDirectionY"][0]
weather_data_dict["Pressure"] = WS_data["Pressure"][0]
```

```python
weather_data_df = pd.DataFrame(weather_data_dict)
weather_data_df.to_parquet(DATA_RAW_DIR / "weather_data.parquet", engine="fastparquet")
```
